```python
import numpy as np
import matplotlib.pyplot as plt

X,Y,array=[],[],[]

with open('1d_team_13_train.txt') as f:
    array = [[float(x) for x in line.split()] for line in f]
    
for x in range(len(array)):
  X.append(array[x][0])
  Y.append(array[x][1])

X=np.array(X)
Y=np.array(Y)
train_data_x = X
train_data_y = Y
train_data = np.asarray(list((zip(train_data_x,train_data_y))))

X,Y,array=[],[],[]

with open('1d_team_13_dev.txt') as f:
    array = [[float(x) for x in line.split()] for line in f]
    
for x in range(len(array)):
  X.append(array[x][0])
  Y.append(array[x][1])

X=np.array(X)
Y=np.array(Y)
test_data_x = X
test_data_y = Y
test_data = np.asarray(list((zip(test_data_x,test_data_y))))

random10 = np.random.choice(train_data.shape[0],200,replace=False)
data = train_data[random10]
Xtemp=data[:,0]
Ytemp=data[:,1]

rand10 = np.random.choice(test_data.shape[0],200,replace=False)
data_test = test_data[rand10]
Xtest=data[:,0]
Ytest=data[:,1]

```

```python
def poly_regression(Xtemp,Ytemp,deg=1):
  A = np.zeros((Xtemp.shape[0],deg+1))
  for i,x in enumerate(data[:,0]):
    A[i] = np.asarray([x**i for i in range(A.shape[1])])
 
  soln = np.dot((np.dot(np.linalg.inv(np.dot(A.T,A)),A.T)),Ytemp)
  y_xw = np.polyval(np.flip(soln),Xtemp)
  error = np.sqrt(np.mean((Ytemp-y_xw)**2))
  return soln,error

def plot_polyreg(Xtemp,Ytemp,soln,deg=1):
  x = np.linspace(0,5,100)
  y_xw = np.polyval(np.flip(soln),x)
  plt.figure(figsize=(9,6))
  plt.grid()
  plt.plot(x ,y_xw,color='red',label='Predicted')
  plt.scatter(Xtemp,Ytemp)
  plt.xlabel(xlabel=r'x',size=15)
  plt.ylabel(ylabel=r'y',size=15)
  plt.title(r'Total datapoints = {}, degree = {}'.format(Xtemp.shape[0],deg),size=15)
  plt.legend()
  plt.savefig('image_'+str(deg)+'.png')
  plt.show()


def poly_reg_error(Xtemp,Ytemp,Xtest,Ytest):
  train_err=[]
  test_err=[]
  for i in range(1,10):
    si,train_error = poly_regression(Xtemp,Ytemp,deg=i)
    y_test = np.polyval(np.flip(si),Xtest)
    test_error = np.sqrt(np.mean((Ytest-y_test)**2))
    train_err.append(train_error)
    test_err.append(test_error)
  
  plt.title(r'RMSE with degree of polynomial',size=15)
  plt.grid()
  plt.plot(np.arange(1,10),train_err,label='train_error',marker='+')
  plt.plot(np.arange(1,10),test_err,label='test_error',marker='o')
  plt.xlabel(xlabel=r'degree of polynomial$\rightarrow$',size=15)
  plt.ylabel(ylabel=r'RMSE$\rightarrow$',size=15)
  plt.legend()
  plt.savefig('rmse.png')
  plt.show()


s1,e1=poly_regression(Xtemp,Ytemp,deg=1)
s3,e3=poly_regression(Xtemp,Ytemp,deg=3)
s5,e5=poly_regression(Xtemp,Ytemp,deg=5)
s6,e6=poly_regression(Xtemp,Ytemp,deg=6)
s7,e7=poly_regression(Xtemp,Ytemp,deg=7)
s8,e8=poly_regression(Xtemp,Ytemp,deg=15)
s9,e9=poly_regression(Xtemp,Ytemp,deg=20)


plot_polyreg(Xtemp,Ytemp,s1,deg=1)
plot_polyreg(Xtemp,Ytemp,s3,deg=3)
plot_polyreg(Xtemp,Ytemp,s5,deg=5)
plot_polyreg(Xtemp,Ytemp,s6,deg=6)
plot_polyreg(Xtemp,Ytemp,s7,deg=7)
plot_polyreg(Xtemp,Ytemp,s7,deg=15)
plot_polyreg(Xtemp,Ytemp,s7,deg=20)

poly_reg_error(data[:,0],data[:,1],data_test[:,0],data_test[:,1])
```

```python
from sklearn.preprocessing import PolynomialFeatures
def poly_2d_calculator(weights,Xtemp,Ytemp,deg):
 ans=0
 
 for i in range(Xtemp.shape[0]):
   for j in range(deg+1):
     for k in range(j+1):
       ans=ans+weights[j]*(Xtemp[i]**k)*(Ytemp**(j-k))
 return ans

def poly_regression_2d(Xtemp,Ytemp,Ztemp,deg=1):
  poly = PolynomialFeatures(deg)
  X=np.asarray(list((zip(Xtemp,Ytemp))))
  A=poly.fit_transform(X)
   
  soln = np.dot((np.dot(np.linalg.inv(np.dot(A.T,A)),A.T)),Ztemp)
  z_xyw = poly_2d_calculator(soln,Xtemp,Ytemp,deg)
  return soln,z_xyw

def plot_polyreg_2d(Xtemp,Ytemp,soln,deg=1):
  x = np.linspace(0,5,100)
  y_xw = np.polyval(np.flip(soln),x)
  plt.figure(figsize=(9,6))
  plt.grid()
  plt.plot(x ,y_xw,color='red',label='Predicted')
  plt.scatter(Xtemp,Ytemp)
  plt.xlabel(xlabel=r'x',size=15)
  plt.ylabel(ylabel=r'y',size=15)
  plt.title(r'Total datapoints = {}, degree = {}'.format(Xtemp.shape[0],deg),size=15)
  plt.legend()
  plt.savefig('image_'+str(deg)+'.png')
  plt.show()

soln,err=poly_regression_2d(X_2d,Y_2d,Z_2d,deg=2)

ax = plt.axes(projection='3d')
ax.scatter3D(X_2d,Y_2d,Z_2d,c=Z_2d, cmap='Greens')

plt.scatter(X_2d,Y_2d,Z_2d)
```

```python
def ridge_regression(data,deg=1, lam=1):
  X = np.zeros((data.shape[0],deg+1)) # n X (d+1) matrix 
  for i,x in enumerate(data[:,0]):
    X[i] = np.asarray([x**i for i in range(deg+1)])
 
  soln = np.dot((np.dot(np.linalg.inv((X.T @ X) + lam*np.identity(deg+1)),X.T)),Ytemp)
  y_xw = np.polyval(np.flip(soln),Xtemp)
  x = np.linspace(0,5,100)
  y_xw = np.polyval(np.flip(soln),x)
  plt.figure(figsize=(9,6))
  plt.title(r'no.of datapoints = {}, degree = {}'.format(data.shape[0],deg),size=15)
  plt.grid()
  plt.plot(x ,y_xw,color='red',label='predicted')
  plt.scatter(Xtemp,Ytemp)
  plt.xlabel(xlabel=r'x',size=15)
  plt.ylabel(ylabel=r'y',size=15)
  plt.legend()
  plt.savefig('reg_image_'+str(deg)+'.png')
  plt.show()

ridge_regression(data, deg=1, lam=0)
ridge_regression(data, deg=2, lam=0)
ridge_regression(data, deg=3, lam=0)
ridge_regression(data, deg=4, lam=0)
ridge_regression(data, deg=5, lam=0)
ridge_regression(data, deg=6, lam=0)
ridge_regression(data, deg=7, lam=0)
ridge_regression(data, deg=10, lam=0)
ridge_regression(data, deg=17, lam=0)
```
